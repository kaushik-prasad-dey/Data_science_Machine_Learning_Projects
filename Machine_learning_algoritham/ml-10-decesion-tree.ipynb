##### What is decesion tree ???
https://www.coursera.org/articles/decision-tree-machine-learning

A decision tree is a supervised learning algorithm that is used for classification and regression modeling. Regression is a method used for predictive modeling, so these trees are used to either classify data or predict what will come next. 

Decision trees look like flowcharts, starting at the root node with a specific question of data, that leads to branches that hold potential answers. The branches then lead to decision (internal) nodes, which ask more questions that lead to more outcomes. This goes on until the data reaches what’s called a terminal (or “leaf”) node and ends.

In machine learning, there are four main methods of training algorithms: supervised, unsupervised, reinforcement learning, and semi-supervised learning. A decision tree helps us visualize how a supervised learning algorithm leads to specific outcomes.

```python
import pandas as pd
df = pd.read_csv("csv/salaries.csv")
df.head()
```

```python
#First we have to drop salary_more_then_100k column
inputs = df.drop('salary_more_then_100k',axis='columns')
```

```python
inputs
```

```python
## this is our target variable
target=df['salary_more_then_100k']
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
### so in this way we can transform our categorical value to number one
# earlier company column was there now we can take one column company_n
# earlier job column was there now we can take one column job_n
# earlier degree column was there now we can take one column degree_n
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_company.fit_transform(inputs['job'])
inputs['degree_n'] = le_company.fit_transform(inputs['degree'])
inputs
```

```python
### when we got all column with its numeric value now we can drop its original column value
inputs_n = inputs.drop(['company','job','degree'],axis='columns')
inputs_n
```

```python
### now we can create our decesion tree model
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
## Now our decesion tree model is created
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[2,1,0]])
# here 2 means working in google
# here 1 means computer programmer
# here 0 means bachelors
```

```python
print("working in google as computer programmer with bachleor degree", model.predict([[2,1,0]]))
print("working in google as computer programmer with master degree", model.predict([[2,1,1]]))
```
