#### Gradient Decent equation calculation & cost function

```python
### MSE = MEAN SQUARE ERROR
### MSE = COST FUNCTION
```

Gradient descent is an algorithm that finds best fit line for given training data set

```python
import numpy as np
```

```python
def gradient_descent(x,y):
    m_curr=b_curr=0
    iterations=1000
    n=len(x)
    learning_rate = 0.5
    for i in range(iterations):
        yp = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y-yp)])
        md=-(2/n) * sum(x*(y-yp))
        bd=-(2/n) * sum(y-yp)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print("m {},b{}, cost{} iteration{}".format(m_curr,b_curr,cost,i))

x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
gradient_descent(x,y)
```

```python
### import the libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import math
```

```python
### create one predict sklearn function and wrap all calling
def predict_using_sklean():
    csvlink ="csv/test_scores.csv"
    df = pd.read_csv(csvlink)
    r = LinearRegression()
    r.fit(df[['math']],df.cs)
    return r.coef_, r.intercept_
```

```python
### Create Gradient Descent function and 
def gradient_descent(x,y):
    m_curr = 0
    b_curr = 0
    iterations = 1000000
    n = len(x)
    learning_rate = 0.0002
    cost_previous = 0

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        #y_predicted = m_curr * X + b_curr
        cost = (1/n)*sum([value**2 for value in (y-y_predicted)])
        #cost = (1/n) *sum([value**2 for value in (y-y_predicted)])
        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        if math.isclose(cost, cost_previous, rel_tol=1e-20):
            break
        cost_previous = cost
        print ("m {}, b {}, cost {}, iteration {}".format(m_curr,b_curr,cost, i))

    return m_curr, b_curr
```

```python
if __name__ == "__main__":
    df = pd.read_csv("csv/test_scores.csv")
    x = np.array(df.math)
    y = np.array(df.cs)

    m, b = gradient_descent(x,y)
    print("Using gradient descent function: Coef {} Intercept {}".format(m, b))

    m_sklearn, b_sklearn = predict_using_sklean()
    print("Using sklearn: Coef {} Intercept {}".format(m_sklearn,b_sklearn))
```
