#### Training & Test Model

```python
import pandas as pd
```

```python
### read the csv file
csv_link = "csv/car_prices_1.csv"
df=pd.read_csv(csv_link)
```

```python
# take the first five rows from the dataset
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# relationship between Mileage and sell price
plt.xlabel('Mileage')
plt.ylabel('Sell Price')
plt.scatter(df.Mileage,df["Sell Price($)"],color='blue')
```

```python
# relationship between age and sell price
plt.xlabel('Age(yrs)')
plt.ylabel('Sell Price')
plt.scatter(df['Age(yrs)'],df["Sell Price($)"],color='green')
```

```python
#now we have to know the X and y ( independent and dependent variable)
# X is independent variable
# y is dependent variable
X = df[['Age(yrs)','Mileage']]
y=df["Sell Price($)"]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
print("length of x",len(X))
print("length of x_test",len(X_test))
print("length of x_train",len(X_train))
```

```python
### Create the linear Regression Model
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(X_train, y_train)
```

```python
## predict the test model
clf.predict(X_test)
```

```python
# Validated the model score
clf.score(X_test,y_test)
```
